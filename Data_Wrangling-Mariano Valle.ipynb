```python
import requests
import json
import pandas as pd
import matplotlib.pyplot as plt
import missingno as msno
```

```python
datos =json.loads(response.text) #nos trae la información como un diccionario de datos
```

## Data Wrangling

```python
df.isnull() #esto se utiliza para la averiguación de nulos
```

```python
msno.bar(df) #no se encuentran nulos
```

```python
df2=df.drop_duplicates()
print(df2.shape) # no se encuentran duplicados
```

```python
#se realiza un renombre de las variables mal escritas

df = df.rename(columns={'area_defunci_n': 'area_defunción'})
df = df.rename(columns={'sitio_defunci_n': 'sitio_defunción'})
df = df.rename(columns={'nombre_instituci_n': 'nombre_institución'})
df = df.rename(columns={'a_o': 'año'})
df = df.rename(columns={'hora_defunci_n': 'hora_defunción'})
df = df.rename(columns={'fecha_defunci_n': 'fecha_defunción'})
df = df.rename(columns={'tipo_defunci_n': 'tipo_defunción'})
df = df.rename(columns={'ultimo_a_o_de_formaci_n': 'ultimo_año_de_formación'})
df = df.rename(columns={'ocupaci_n_fallecido': 'ocupación_fallecido'})
df = df.rename(columns={'pa_s_residencia': 'país_residencia'})
df = df.rename(columns={'r_gimen_seguridad': 'régimen_seguridad'})
df = df.rename(columns={'pertenencia_tnica': 'pertenencia_étnica'})
```

```python
df.columns # Se comprueba el cambio de nombre de las variables
```

```python
# Se Convertierte la columna de fecha_defunción al tipo de dato datetime

df['fecha_defunción'] = pd.to_datetime(df['fecha_defunción'])

# Crear nuevas columnas para mes y año
df['mes'] = df['fecha_defunción'].dt.month
df['año'] = df['fecha_defunción'].dt.year

# Muestra las primeras filas del DataFrame resultante
print(df[['fecha_defunción', 'mes', 'año']].head())
```

```python
# Se realzia un histograma para ver los datos de la varible

plt.figure(figsize=(10, 6))
df['fecha_defunción'].hist(bins=30, color='skyblue', edgecolor='black')

plt.title('Histograma de fechas de defunción')
plt.xlabel('Fecha de Defunción')
plt.ylabel('Frecuencia')

plt.show()
```

```python
# se realiza un conteo de datos de la variable "sexo_fallecido"
sexo_counts = df['sexo_fallecido'].value_counts()
print(sexo_counts)
```

```python
# se grafica el conteo de la variable "sexo_fallecido"

sexo_counts.plot(kind='bar', color=['blue', 'pink', 'gray'])
plt.title('Sexo de victimas fallecidas')
plt.xlabel('Sexo ')
plt.ylabel('Número de Fallecidos')
plt.show()
```

```python
# se realiza un conteo de los barrios
barrio_counts = df['barrio'].value_counts()
print(barrio_counts)
```

```python
# se realiza una visualización de todo el conteo
pd.set_option('display.max_rows', None)
print(barrio_counts)
```

```python
# Ya que son muchos datos para incluir en un grafico, solo vamos a incluir un top 10
top_10_barrios = barrio_counts.head(10)
print("\nTop 10 de Barrios:")
print(top_10_barrios)
```

```python
# se grafica el top 10
plt.figure(figsize=(10, 6))
top_10_barrios.plot(kind='bar', color='skyblue')
plt.title('Top 10 de Barrios con Mayores fallecidos')
plt.xlabel('Barrios')
plt.ylabel('Número de Fallecidos')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# se realiza un conteo del lugar de las defunciones
lugardefunciones_counts = df['sitio_defunción'].value_counts()
print(lugardefunciones_counts)
```

```python
lugardefunciones_counts.plot(kind='bar', color=["blue","brown","gray","red"])
plt.title("Lugar de defunciones")
plt.xlabel("Ubicación de Defunción")
plt.ylabel("Cantidad de defunciones")
plt.show()
```
